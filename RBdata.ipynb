```python
#import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#path to the csv file
rbdatapath = "Resources/NFLRBDATA.csv"
```

```python
#read csv file
rb_data_df=pd.read_csv(rbdatapath)
```

```python
#display 5 rows of csv data
rb_data_df.head()
```

```python
#total number of players in the position (running back)
total_players= len(rb_data_df)
total_players
```

```python
#eliminate rows for players who played less than 1 season
condensed=rb_data_df.loc[rb_data_df["num_seasons_played"]>1]
condensed_rbdata= condensed.loc[condensed["To"]!=2019]
condensed_rbdata
```

```python
#calc avg length of career for players
avg_career= condensed_rbdata["num_seasons_played"].mean()
print(f"Average length of career for a Running Back- {avg_career}")
stdev_rb= condensed_rbdata["num_seasons_played"].std()
print(f"Standard deviation for Running Back {stdev_rb}")
median_rb=condensed_rbdata["num_seasons_played"].median()
print(f"Median for Running Back {median_rb}")
```

```python
quartile1= np.percentile(condensed_rbdata['num_seasons_played'],25)
quartile3= np.percentile(condensed_rbdata['num_seasons_played'],75)
rb_iqr=(quartile3-quartile1)
upper_bound=quartile1-(1.5*rb_iqr)
lower_bound=quartile3+(1.5*rb_iqr)
print(f"First Quartile : {quartile1}")
print(f"Third Quartile: {quartile3}")
print(f"Inter Quartile Range: {rb_iqr}")
print(f"Upper Boundary: {upper_bound}   Lower Boundary: {lower_bound}")
```
